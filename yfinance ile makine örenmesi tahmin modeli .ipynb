```python
import yfinance as yf
```

```python
altın = yf.Ticker("GC=F") # Altın verilerini almak için yfinance kütüphanesini içe aktarırız
```

```python
altın = altın.history(period="max") # Altın verilerini geçmişe yönelik almak için history fonksiyonunu kullanırız
```

```python
altın # Altın verilerini ekrana yazdırır
```

```python
altın.index # Altın verilerinin indeksini ekrana yazdırır
```

```python
altın.plot.line(y="Close", use_index = True) # Altın verilerini "Close" sütununu kullanarak çizgi grafiği olarak çizer
```

```python
# Altın verilerinden "Dividends" sütununu siler
# Altın verilerinden "Stock Splits" sütununu siler
del altın["Dividends"]
del altın["Stock Splits"]
```

```python
altın["Yarın"] = altın["Close"].shift(-1) # Altın verilerine "Yarın" adında yeni bir sütun ekler ve "Close" sütununu bir birim yukarı kaydırır
```

```python
altın
```

```python
# Altın verilerine "Hedef" adında yeni bir sütun ekler ve "Yarın" sütunu ile "Close" sütununu karşılaştırarak 1 veya 0 değerleri atar
altın["Hedef"] = (altın["Yarın"] > altın["Close"]).astype(int) 
```

```python
altın
```

```python
# Random Forest sınıflandırma modelini oluşturur
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
# Eğitim verilerini oluşturur
train = altın.iloc[:-100]
# Test verilerini oluşturur
test = altın.iloc[-100:]
# Kullanılacak öngörücü özellikleri belirler
ongoruculer = ["Close","Volume", "Open" , "High" , "Low"]
# Modeli eğitir
model.fit(train[ongoruculer], train["Hedef"])
```

```python
from sklearn.metrics import precision_score
# Test verileri üzerinde tahmin yapar
preds = model.predict(test[ongoruculer])
```

```python
import pandas as pd 
# Tahminleri bir pandas Serisi olarak oluşturur
preds = pd.Series(preds, index=test.index)
```

```python
# Doğruluk değerini hesaplar
precision_score(test["Hedef"],preds)
```

```python
# Test verilerinin "Hedef" ve "Predictions" sütunlarını birleştirir

combined = pd.concat([test["Hedef"], preds] , axis=1)
```

```python
# Birleştirilen verileri çizgi grafiği olarak çizer

combined.plot()
```

```python
# Tahmin fonksiyonunu tanımlar

def predict(train, test, ongoruculer, model):
    model.fit(train[ongoruculer], train["Hedef"])
    preds = model.predict(test[ongoruculer])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Hedef"], preds] , axis=1)
    return combined
```

```python
# Geriye dönük test fonksiyonunu tanımlar

def backtest(data , model, ongoruculer, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train ,test, ongoruculer, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
        
```

```python
# Altın verileri üzerinde geriye dönük test yapar

predictions = backtest(altın, model, ongoruculer)
```

```python
# "Predictions" sütunundaki tahmin değerlerinin sıklıklarını hesaplar

predictions["Predictions"].value_counts()
```

```python
# Doğruluk değerini hesaplar

precision_score(predictions["Hedef"],predictions["Predictions"])
```

```python
# "Hedef" sütunundaki değerlerin yüzdesel dağılımını hesaplar

predictions["Hedef"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

# Belirli zaman aralıkları için yeni özellikler oluşturur

for horizon in horizons:
    rolling_averages = altın.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    altın[ratio_column] = altın["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    altın[trend_column] = altın.shift(1).rolling(horizon).sum()["Hedef"]
    
    new_predictors += [ratio_column, trend_column]
    
```

```python
# Eksik verilere sahip satırları çıkarır

altın = altın.dropna()
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
# Tahmin fonksiyonunu yeniden tanımlar

def predict(train, test, ongoruculer, model):
    model.fit(train[ongoruculer], train["Hedef"])
    preds = model.predict_proba(test[ongoruculer])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Hedef"], preds] , axis=1)
    return combined
```

```python
# Altın verileri üzerinde geriye dönük test yapar

predictions = backtest(altın, model, new_predictors)
```

```python
# "Predictions" sütunundaki tahmin değerlerinin sıklıklarını hesaplar

predictions["Predictions"].value_counts()
```

```python
# Doğruluk değerini hesaplar

precision_score(predictions["Hedef"], predictions["Predictions"])
```
